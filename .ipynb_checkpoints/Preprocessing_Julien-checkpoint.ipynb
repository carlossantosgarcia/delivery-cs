```python
import pandas as pd
import os
import re


#Nous ouvrons le dataset
df_ACE = pd.read_csv(os.path.join(r"comptages-routiers-permanents_ACE.csv"), sep=";")
#print(df_ACE.head())

def clean_date(date):
    
    date = re.sub('T', ' ', date)
    date=date[:-6]
    return date
compte_1=df_ACE.groupby("Etat arc")["Date et heure de comptage"].nunique()
compte_2=df_ACE.groupby("Identifiant noeud aval")["Date et heure de comptage"].nunique()
compte_3=df_ACE.groupby("Identifiant noeud amont")["Date et heure de comptage"].nunique()
compte_4=df_ACE.groupby("Libelle")["Date et heure de comptage"].nunique()
print(compte_1)
print(compte_2)
print(compte_3)
print(compte_4)
df_ACE['Date et heure de comptage'] = df_ACE['Date et heure de comptage'].apply(lambda s : clean_date(s))
print(df_ACE['Date et heure de comptage'])
#Nous adaptons notre features "Date" au format d'une série temporelle
df_ACE['Date et heure de comptage']= pd.to_datetime(df_ACE["Date et heure de comptage"], format='%Y-%m-%d %H:%M:%S')
df_ACE.sort_values("Date et heure de comptage",inplace = True)
#Nous indiquons que la série temporelle est indexée selon la date

df_ACE=df_ACE.set_index('Date et heure de comptage')

#Nous créons plusieurs colonnes permettant de situer la donnée par jour/mois/année. Cela nous servira pa la suite pour traiter
#la données selon différentes échelles temporelles.
df_ACE['Year'] = df_ACE.index.year
df_ACE['Month'] = df_ACE.index.month
df_ACE['Weekday Name'] = df_ACE.index.weekday
df_ACE['Hour'] = df_ACE.index.hour

print(df_ACE.index)
```

```python
df_ACE.drop(columns=["Libelle noeud aval","Libelle noeud amont","Libelle","Identifiant noeud aval","Identifiant noeud amont"], inplace = True)
print(df_ACE.head())
```

```python
#Nous décomposons ici la série temporelle en isolant sa tendance, sa saisonnalité et ces résidus
import statsmodels.api as sm
import matplotlib
decomposition_1= sm.tsa.seasonal_decompose(df_ACE["Débit horaire"], model = 'additive')
decomposition_2= sm.tsa.seasonal_decompose(df_ACE["Taux d'occupation"], model = 'additive')
matplotlib.rcParams['figure.figsize']= [9.0,5.0]
decomposition_1.plot()
decomposition_2.plot()
plt.show()
```

```python
#Nous repésentons désormais le graphe d'autocorrélation de la série temporelle. Les lignes en poitillés délimite les valeurs 
#non significatives.
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(df_ACE["Débit horaire"])
```
