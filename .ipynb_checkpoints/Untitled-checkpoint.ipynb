```python
import os
import datetime

from functools import reduce

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('data/champs-elysees-cleaned.pkl')
df.head()
```

```python
df['Date'].iloc[0]
```

```python
df['Temps avant les prochaines vacances scolaires'].iloc[0]
```

```python
df_daily = pd.read_pickle('data/daily_weather_data_from_2009_to_present.pkl')
df_daily.head()
```

```python
df_hourly = pd.read_pickle('data/hourly_weather_data_from_2009_to_present.pkl')
df_hourly.head()
```

```python
df = df.merge(df_hourly, left_on='Date et heure de comptage', right_on='datetime', how='left').drop(columns=['datetime'])
```

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(230, 20, as_cmap=True)
with sns.axes_style("white"):
    plt.figure(figsize=(10,10))
    ax = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, square=True)
```
