```python
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.feature_selection import RFECV
```

```python
df = pd.read_pickle('data/df_concat_train.pkl')
df.head()
```

```python
cols_prediction = ['champs-elysees_Débit horaire', "champs-elysees_Taux d'occupation"]
y = df[cols_prediction]
y.head()
```

```python
X = df.drop(columns=df.columns[1:14])
X.head()
```

```python
estimator = SGDRegressor(max_iter=1000, tol=1e-3)
```

```python
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(X, y)
```

```python
selector.support_
```
